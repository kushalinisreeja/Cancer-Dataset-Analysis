```python
#creats the new column by performing the arithmetic operation
df['treatment_duration'] = (df['end_treatment_date'] - df['diagnosis_date']).dt.days
df
```

```python
# Count of male and female patients
gender_counts = df['gender'].value_counts()

# Average age for each gender
average_age_by_gender = df.groupby('gender')['age'].mean()

# Display results
print("Number of male and female patients:")
print(gender_counts)

print("\nAverage age for each gender:")
print(average_age_by_gender)
```

```python
# Count survival outcome by gender
survival_by_gender = df.groupby('gender')['survived'].value_counts(normalize=True).unstack()

# Count of treatment types by gender
treatment_by_gender = df.groupby('gender')['treatment_type'].value_counts(normalize=True).unstack()

# Count of cancer stages by gender
stage_by_gender = df.groupby('gender')['cancer_stage'].value_counts(normalize=True).unstack()

# Display as percentages
print("Survival rate by gender (%):")
print((survival_by_gender * 100).round(2))

print("Treatment type distribution by gender (%):")
print((treatment_by_gender * 100).round(2))

print("Cancer stage distribution at diagnosis by gender (%):")
print((stage_by_gender * 100).round(2))
```

```python
#counts the patients per country
patients_per_country = df['country'].value_counts()
print("Top 5 countries with highest number of patients:")
print(patients_per_country.head(5))

#gets the survival rate
survival_rate_by_country = df.groupby('country')['survived'].mean()

#gets the survival rate of top 5 countries
top5_by_survival = survival_rate_by_country.sort_values(ascending=False).head(5)
print("Top 5 countries which are having best survival rate (%):")
print((top5_by_survival * 100).round(2))
```

```python
#gets average by different countries
avg_age_by_country = df.groupby('country')['age'].mean().round(2)
print("Average age across different countries:")
print(avg_age_by_country)

#gets the average age different cancer stages
avg_age_by_stage = df.groupby('cancer_stage')['age'].mean().round(2)
print("Average age by cancer stages:")
print(avg_age_by_stage)

# Count treatment types across age groups
treatment_by_age = df.groupby('age_group',observed=True)['treatment_type'].value_counts(normalize=True).unstack()
print("Treatment types across different age groups (%):")
print((treatment_by_age * 100).round(2))
```

```python
#gets the survival rate across different stages
survival_by_stages = df.groupby('cancer_stage')['survived'].value_counts(normalize=True).unstack()
print("Survival rates across different cancer stages (%):")
print((survival_by_stages * 100).round(2))
```

```python
# Death rate = 1 - survival rate
death_rate_by_stage = 1 - df.groupby('cancer_stage')['survived'].mean()

# Convert to percentage and sort
death_rate_by_stage = (death_rate_by_stage * 100).round(2).sort_values(ascending=False)

print("Death rate (%) by cancer stage:")
print(death_rate_by_stage)

# Highest death rate stage
highest_death_stage = death_rate_by_stage.idxmax()
print(f"\nCancer stage with the highest death rate: {highest_death_stage}")
```

```python
survival_by_age_group = df.groupby('age_group',observed=True)['survived'].mean().round(3)
print("Survival probability by age group:")
print(survival_by_age_group)
```

```python
#creating a mapping dictionary
stage_map = {
    'Stage I': 'Early',
    'Stage II': 'Early',
    'Stage III': 'Late',
    'Stage IV': 'Late'
}

#map cancer stages to Early/Late groups
df['stage_group'] = df['cancer_stage'].map(stage_map)

#calculate survival rate for each group
survival_by_stage_group = df.groupby('stage_group')['survived'].mean().round(3)
survival_percent = (survival_by_stage_group * 100).round(2)
print("Survival rate (%) by stage group:")
print(survival_percent)
```

```python
import matplotlib.pyplot as plt

survival_percent.plot(kind='bar', color=['green', 'red'])
plt.title('Survival Rate: Early vs Late Stage Diagnosis')
plt.ylabel('Survival Rate (%)')
plt.xticks(rotation=0)
plt.ylim(0, 100)
plt.show()
```

```python
stage_distribution = df['cancer_stage'].value_counts()
print("Cancer stage distribution (counts):")
print(stage_distribution)
```

```python
import matplotlib.pyplot as plt

stage_distribution.plot(kind='bar', color='skyblue')
plt.title('Cancer Stage Distribution at Diagnosis')
plt.xlabel('Cancer Stage')
plt.ylabel('Number of Patients')
plt.xticks(rotation=0)
plt.show()
```

```python
# Group by cancer stage and treatment type, then calculate survival rate
survival_rates = df.groupby(['cancer_stage', 'treatment_type'])['survived'].mean().reset_index()

# Round the survival rates for better readability
survival_rates['survival_rate'] = (survival_rates['survived'] * 100).round(2)

# Drop the old 'survived' column
survival_rates = survival_rates.drop(columns='survived')

# For each cancer stage, find the treatment with the highest survival rate
best_treatment = survival_rates.sort_values('survival_rate', ascending=False).groupby('cancer_stage').first()

print("Best treatment type for each cancer stage:")
print(best_treatment)
```

```python
# Count of each treatment type
most_common_treatment = df['treatment_type'].value_counts()
print("Most commonly used treatment types (overall):")
print(most_common_treatment)

# Count how many patients received each treatment per stage
treatment_by_stage = df.groupby('cancer_stage')['treatment_type'].value_counts().unstack(fill_value=0)
print("Treatment counts by cancer stage:")
print(treatment_by_stage)
```

```python
import matplotlib.pyplot as plt

treatment_by_stage.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Treatment Distribution Across Cancer Stages')
plt.ylabel('Number of Patients')
plt.xlabel('Cancer Stage')
plt.legend(title='Treatment Type')
plt.tight_layout()
plt.show()
```

```python
# Map treatment types to categories
df['treatment_category'] = df['treatment_type'].map(
    lambda x: 'Combination' if x == 'Combined' else 'Single'
)

# Group by treatment category and calculate survival rate
survival_by_treatment_type = df.groupby('treatment_category')['survived'].mean().round(3)

# Convert to percentage (optional)
survival_percentage = (survival_by_treatment_type * 100).round(2)

print("Survival rate (%) by treatment type:")
print(survival_percentage)
```

```python
#finds the average treatment duration
avg_duration = df['treatment_duration'].mean().round(2)
print(f"Average treatment duration: {avg_duration} days")


#spliting patients into two groups based on the average duration
df['duration_group'] = df['treatment_duration'].map(
    lambda x: 'Short' if x < avg_duration else 'Long'
)


#compare survival rate forr each group
survival_by_duration = df.groupby('duration_group')['survived'].mean() * 100
print("Survival rate (%) by treatment duration group:")
print(survival_by_duration.round(2))
```

```python
# Group by country and calculate average duration and survival
country_stats = df.groupby('country').agg(
    avg_duration=('treatment_duration', 'mean'),
    survival_rate=('survived', 'mean')
)

# Convert survival rate to percentage and round both
country_stats['survival_rate'] = (country_stats['survival_rate'] * 100).round(2)
country_stats['avg_duration'] = country_stats['avg_duration'].round(1)

# Show countries sorted by highest survival
print(country_stats.sort_values('survival_rate', ascending=False))
```

```python
# Calculate survival rate (%) by smoking status
smoking_survival = df.groupby('smoking_status')['survived'].mean() * 100

print("Survival rate by smoking status (%):")
print(smoking_survival.round(2))


# Get distribution (%) of cancer stages within each smoking group
stage_by_smoking = df.groupby('smoking_status')['cancer_stage'].value_counts(normalize=True).unstack()

print("Cancer stage distribution by smoking status (%):")
print((stage_by_smoking * 100).round(2))
```

```python
#creates bmi categories
df['bmi_category'] = pd.cut(df['bmi'], bins=[0, 19, 25, 30, 100], labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

# Group by BMI category and calculate average survival rate
bmi_survival = df.groupby('bmi_category',observed=True)['survived'].mean() * 100

print("Survival rate by BMI category (%):")
print(bmi_survival.round(2))


```

```python
import matplotlib.pyplot as plt

bmi_survival.plot(kind='bar', color='mediumseagreen')
plt.title('Survival Rate by BMI Category')
plt.ylabel('Survival Rate (%)')
plt.xlabel('BMI Category')
plt.xticks(rotation=0)
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

```python
comorbidities = ['asthma', 'hypertension', 'cirrhosis']

#Check how cancer stages are distributed among patients with and without each condition.
for condition in comorbidities:
    print(f"\nCancer stage distribution for {condition}:")
    stage_dist = df.groupby(condition)['cancer_stage'].value_counts(normalize=True).unstack()
    print((stage_dist * 100).round(2))

#compares survival(%) for patients with and withut each condition.
for condition in comorbidities:
    print(f"\nSurvival rate by {condition}:")
    survival_rate = df.groupby(condition)['survived'].mean() * 100
    print(survival_rate.round(2))
```

```python
# Sum the binary comorbidity columns row-wise 
df['comorbidity_count'] = df[comorbidities].sum(axis=1)


# Create comorbidity level group
df['comorbidity_level'] = df['comorbidity_count'].map(
    lambda x: 'None' if x == 0 else '1 Condition' if x == 1 else '2+ Conditions'
)

survival_by_comorbidity = df.groupby('comorbidity_level')['survived'].mean() * 100

print("Survival rate by comorbidity level:")
print(survival_by_comorbidity.round(2))
```

```python
#gets the average age at diagnosis by family history
age_by_history = df.groupby('family_history')['age'].mean()
print("Average age at diagnosis:")
print(age_by_history.round(2))

#survival rate by family history
survival_by_history = df.groupby('family_history')['survived'].mean() * 100
print("\nSurvival rate by family history (%):")
print(survival_by_history.round(2))
```

```python
import matplotlib.pyplot as plt

# Bar plot for average diagnosis age
age_by_history.plot(kind='bar', color='orange')
plt.title("Average Age at Diagnosis by Family History")
plt.ylabel("Age")
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Calculate percentage distribution of stages within each group
stage_dist = df.groupby('family_history')['cancer_stage'].value_counts(normalize=True).unstack()

# Convert to percentage
stage_percent = (stage_dist * 100).round(2)

print("Cancer stage distribution (%) by family history:")
print(stage_percent)
```

```python
stage_percent.T.plot(kind='bar', figsize=(8, 5))
plt.title("Cancer Stage Distribution by Family History")
plt.ylabel("Percentage of Patients")
plt.xlabel("Cancer Stage")
plt.xticks(rotation=0)
plt.legend(title='Family History', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
# Survival rate for patients with vs without other cancers
survival_other_cancer = df.groupby('other_cancer')['survived'].mean() * 100
print("Survival rate by presence of other cancers:")
print(survival_other_cancer.round(2))


# Group by other_cancer + treatment_type, then calculate survival %
treatment_effectiveness = (
    df.groupby(['other_cancer', 'treatment_type'])['survived']
    .mean()
    .unstack()
    .round(2) * 100
)

print("Survival rate (%) by treatment and other cancer presence:")
print(treatment_effectiveness)
```
